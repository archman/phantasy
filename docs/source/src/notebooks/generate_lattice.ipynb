# Generate FLAME lattice file from machine
This notebook shows how to generate the lattice file for FLAME model engine, with the device settings from accelerators (virtual or real accelerators)

```python
from phantasy import MachinePortal
from phantasy import generate_settings

from phantasy.tools import plot_orbit
from phantasy import disable_warnings
disable_warnings()
```

```python
mp = MachinePortal(machine='LEBT')
lat = mp.work_lattice_conf
snpfile = './Ar_LEBT_to_MEBT_20180321.snp'
settings_read = generate_settings(snpfile=snpfile, lattice=lat, only_physics=True)
```

```python
latfile1, fm1 = lat.run()
```

```python
data1 = plot_orbit(('pos', 'xcen'), ('pos', 'ycen'), flame_model=fm1)
```

```python
lat.settings = settings_read
```

```python
latfile2, fm2 = lat.run()
```

```python
data2 = plot_orbit(('pos', 'xcen'), ('pos', 'ycen'), flame_model=fm2)
```
